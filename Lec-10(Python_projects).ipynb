```python
# Project-I : Guess the number game:(My Approach)

import random
target1=random.randint(1,10)
target2=random.randint(1,100)
target3=random.randint(1,1000)

ans = "Y"                                             # we have to initialize ans with the value that will make the game run, otherwise it will throw error

while True:
    if ans=="Y":

        print("Welcome to Number Guessing Game")
        print("Please Select Your Difficulty","\n(i) Lvl.I: Between 1 to 10. Difficulty Multiplier = 1","\n(ii) Lvl.II: Between 1 to 100. Difficulty Multiplier = 10","\n(iii) Lvl.III: Between 1 to 1000. Difficulty Multiplier = 100")
        print("Enter your options in 1 , 2 , 3")

        opt=int(input("Enter Option: "))


        if opt==1:
            print("Lvl.I selected")
            print("Now you have to guess the number between 1 to 10")
            target = target1
            att = 0
            while True:
                n=int(input("Enter number: "))
                att+=1
                if n==target:
                    print("Correct Guess!!","Congratulation")
                    break
                elif (n>target and n<=10):
                    print("Too High")
                elif (0<n<target):
                    print("Too Low")
                else:
                    print("Invalid Guess, Guess number from 1 to 10")
            print("Game Over, You Won!!")
            print("Attempts Taken-> ",att)

        elif opt==2:
            print("Lvl.II selected")
            print("Now you have to guess the number between 1 to 100")
            target = target1
            atte = 0
            while True:
                n=int(input("Enter number: "))
                atte+=1
                if n==target:
                    print("Correct Guess!!","Congratulation")
                    break
                elif (n>target and n<=100):
                    print("Too High")
                elif (0<n<target):
                    print("Too Low")
                else:
                    print("Invalid Guess, Guess number from 1 to 100")
            print("Game Over, You Won!!")
            print("Attempts Taken-> ",atte)

        elif opt==3:
            print("Lvl.III selected")
            print("Now you have to guess the number between 1 to 1000")
            target = target1
            atte = 0
            while True:
                n=int(input("Enter number: "))
                atte+=1
                if n==target:
                    print("Correct Guess!!","Congratulation")
                    break
                elif (n>target and n<=1000):
                    print("Too High")
                elif (0<n<target):
                    print("Too Low")
                else:
                    print("Invalid Guess, Guess number from 1 to 1000")
            print("Game Over, You Won!!")
            print("Attempts Taken-> ",atte)

        print("Type 'Y' to play again and 'N' to exit ")
        ans=input("Do you want to play again?(Y/N) : ").upper()

    elif ans=="N":
        print("Exited from game sucessfully, Thanks for playing!! ")
        break
    else:
        print("Invalid Answer, Answer in Y or N only")
        ans=input("Do you want to play again?(Y/N) : ").upper()                   # if we do not do this infinite loop will occur

```

```python
# Alt Approah (not mine):

# List comprehension : A list comprehension in Python is a concise, elegant way to create a new list from an existing iterable (like a list, tuple, or range) in a single line of code
# new_list = [expression for item in iterable if condition]

# syntax here => [function for i in range(parameter)]      ; so we create this to print password without creating a long loop

import random
import string

values = string.ascii_letters + string.digits + string.punctuation

print("===============WELCOME TO RANDOM PASSWORD GENERATOR================")
print("The application will create a strong, random password for you based on your preffered length")
print("Please Enter a note to remember your password")
# x=input("Enter Note: ")
print("Please enter the number of characters you want in your password(length of the password): ")
pass_len=int(input("Enter number of characters: "))

result = [random.choice(values) for i in range(pass_len)]
# print(result)                                             # this will print all the characters in of the password as sting in list ['4', 'k', ']', '`', '4', '$', '*', 'v', 'O', 'r', '+', '6']
                                                            # if we want them as a single string character wee have to use the .join function

password="".join(result)                                    #.join basically concatenates all the string chracters ; "_".join at place o _ we can add anything we want between each string character of result list

print("Your Unique",pass_len,"charactered strong password is--> ",password)
```
